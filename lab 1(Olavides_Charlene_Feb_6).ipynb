```python
#1. See the list below. Please write a program that will tell you what the longest word is in the list. Package your program inside of a function.  
my_list = ["apple", "chestnut", "gargoyle", "pandas", "sheep", "raptor"]

def check(): 
    my_list.sort(key = len)#this will sort the list from least letters to most letters, https://www.geeksforgeeks.org/python-how-to-sort-a-list-of-strings/
    longest_word = my_list[5] #since there are six words, the 5th indexed word will be the longest
    print('The longest word is ' + longest_word + '.')
    
check()
```
