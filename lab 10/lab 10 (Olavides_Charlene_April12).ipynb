```python
from tkinter import *
import requests
import random
import json

url = requests.get("https://api.quotable.io/authors?page=1&limit=150")
data = url.json()
a = data['results']

list_people = []
for i in a:
    name = i['name']
    list_people.append(name)
    
print(list_people)
```

```python
random = random.choices(list_people,k=8)
random
```

```python
root = Tk()
root.title("Quotes")
root.geometry("900x400")

def chosen_person():
    person = clicked.get()
    result_label = Label(root, text=f"You chose {person}")
    result_label.pack()
    
    url2 = f"https://api.quotable.io/random?author={person}"
    data = requests.get(url2)
    quotes = data.json()
    quote = quotes['content']
        
    quote_label = Label(root, text=quote)
    quote_label.pack()
       

clicked = StringVar()
clicked.set("Select a person, get a quote from them")

dropdown = OptionMenu(root, clicked, *random)
choose_button = Button(root, text="Click to choose", command=chosen_person)
#I couldn't actually figure out how to reset the menu

dropdown.pack()
choose_button.pack()

root.mainloop()
```

```python
root = Tk()
root.title("Weather Forecast")
root.geometry("700x700")

lat = "40.7128"
lon = "-74.0060"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
                      
json_file = genius.json()
json_file

forecast=json_file['properties']['forecast']

new_request = requests.get(forecast)

json = new_request.json()
json
    
w = json['properties']['periods']
w
```

```python
label_header = Label(root, text='7-day forecast: Hoboken, NJ', font=('Arial',20))
label_header.pack()

canvas = Canvas(root, width = 300, height = 300) #new widget      
canvas.pack()      
img = PhotoImage(file="wpic.png")      
canvas.create_image(20,20, anchor=NW, image=img) 

for i in w:
    day = i['name']
    temperature = i['temperature']
    
    label_day = Label(root, text=f"{day}: {temperature}", font=('Arial',15))
    label_day.pack()
```

```python
root.mainloop()
```
