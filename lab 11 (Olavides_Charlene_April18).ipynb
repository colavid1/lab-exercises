```python
import random
from tkinter import *
import requests

root = Tk()
root.title('Trivia')
root.geometry("800x900")


url = requests.get(f"https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy")
data = url.json()

canvas = Canvas(root, width=700, height=300, bg="LightSkyBlue1")
canvas.place(relx=.5, rely=.5,anchor= CENTER)

game_display = canvas.create_text(200, 75, text="", fill="black", font=('Helvetica 15 bold'))
feedback = Label(root, text="", pady=10, font=("arial", 15, "bold"))

user_answer = StringVar()
user_answer.set("")
score = 0
questions_correct = 0
def execute():
        
    def get_questions():
        global list_questions
        list_questions = []
        for i in data:
            question = i['question']
            list_questions.append(question)
    
    def pull_answers():
        global randomq
        randomq = random.choice(list_questions)
    
        global option_list
        option_list = []
    
        for i in data:
            if randomq == i['question']:
                global correct_ans
                correct_ans = i['correctAnswer']
                option_list.append(correct_ans)
                for incor in i["incorrectAnswers"]:
                    option_list.append(incor)
                list_questions.remove(randomq)
                
        random.shuffle(option_list)
    
    def ask_quest():
         canvas.itemconfig(game_display, text=randomq, width=300)
    
    def gen_radiobuttons():
        y_pos = 100
        for i in option_list:
            global rad_btn
            rad_btn = Radiobutton(canvas, text=i, variable=user_answer, command=eval_score, value=i, font=('Helvetica 10 bold'))
            rad_btn.place(x=50, y=y_pos)
            y_pos+=20        
        
    def eval_score():
        ans = user_answer.get()
        for i in data:
            if randomq == i['question']:
                if ans == correct_ans:
                    feedback['text'] = 'Correct!'
                    feedback.place(x=350,y=450)
                    global score
                    global questions_correct
                    score+=10
                    questions_correct+=1
                else:
                    feedback['text'] = 'Incorrect'
                    feedback.place(x=350,y=450)  
        return score
        return questions_correct
            
    def finished_questions():
        if list_questions == []:
            for i in option_list:
                rad_btn.destroy()
            canvas.itemconfig(game_display, text=f'You have gone through all questions!You earned {score} points! You answered {questions_correct} questions correctly', width=300)
            
    def nextques():
        try:
            for i in option_list:
                rad_btn.destroy() #I tried to use this to clear the radio buttons so the answers wouldn't overlap with each new questions but I couldn't get it to work. I think it has something to do with how this program processes each new question and answer set after pressing the Next button
            pull_answers()
            ask_quest()
            gen_radiobuttons()
            eval_score()
        except:
            finished_questions()
    
    def gen_nextqButton():
        nextButton = Button(canvas, text="Next", font=("arial", 10), command=nextques)
        nextButton.place(x=600,y=250)
    
    get_questions()
    pull_answers()
    ask_quest()
    gen_radiobuttons()
    eval_score()
    finished_questions()
    nextques()
    gen_nextqButton()
    
        
execute()
root.mainloop()                
```
